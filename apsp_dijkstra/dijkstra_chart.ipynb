```python
import matplotlib.pyplot as plt

gpu_times = []
cpu_times = []
sizes = []

with open('dijkstra_results.log', 'r') as file:
    for line in file:
        parts = line.split(' ')
        time = float(parts[5])
        size = int(parts[9])
        
        if 'GPU' in line:
            gpu_times.append(time)
        elif 'CPU' in line:
            cpu_times.append(time)
        
        sizes.append(size)

gpu_sizes = sizes[::2]
cpu_sizes = sizes[1::2]  

plt.figure(figsize=(10, 6))

plt.plot(gpu_sizes, gpu_times, marker='o', label='Dijkstra GPU', color='green')
plt.plot(cpu_sizes, cpu_times, marker='o', label='Dijkstra CPU', color = 'orange')

plt.title('Dijkstra GPU CUDA X CPU SERIAL')
plt.xlabel('Size')
plt.ylabel('Execution Time (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
